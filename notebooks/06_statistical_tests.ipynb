```python
# 06_statistical_tests.ipynb
# =============================================================================
# Personalized Health Intervention Pathways for Comorbid Policyholders:
# Loss Ratio Management via LLM Guardrails and Counterfactual Explanations
# -----------------------------------------------------------------------------
# Notebook 06 — Statistical Tests & Guardrail Qualitative Comparison
#
# Contents:
#   1. Kruskal-Wallis test: guardrail range heterogeneity across groups
#   2. Post-hoc pairwise tests (Mann-Whitney U + Bonferroni)
#   3. Fallback rate analysis (actual DiCE search results)
#   4. Numerical hallucination comparison table
#      (Pure DiCE vs. Guardrail DiCE — qualitative ablation)
#
# Input  : ../results/tables/agent_config.pkl
#          ../results/tables/df_final.pkl
#          ../results/tables/model_{group}.pkl
#          ../results/tables/guardrail_ranges.json
#          ../results/tables/guardrail_descriptive.csv
# Output : ../results/tables/kruskal_wallis_results.csv
#          ../results/tables/dunn_posthoc.csv
#          ../results/tables/fallback_analysis.csv
#          ../results/tables/hallucination_comparison.csv
#          ../results/figures/figure_hallucination_comparison.{png|pdf|tiff}
# =============================================================================

# %%
# ## 0. Import Libraries

import json
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import dice_ml

from scipy.stats import kruskal, mannwhitneyu, chi2_contingency
from itertools import combinations

warnings.filterwarnings('ignore')

SAVE_FIG_KWARGS = dict(dpi=600, bbox_inches='tight', facecolor='white')

def save_fig(fig, name: str):
    for ext in ['png', 'pdf', 'tiff']:
        fig.savefig(f'../results/figures/{name}.{ext}',
                    format=ext, **SAVE_FIG_KWARGS)
    print(f"  Saved: {name} (PNG/PDF/TIFF 600 dpi)")

# %%
# ## 1. Load Data

agent_config = joblib.load('../results/tables/agent_config.pkl')
df_final     = joblib.load('../results/tables/df_final.pkl')

X_FEATURES      = agent_config['X_features']
VARY_FEATURES   = agent_config['vary_features']
FIXED_FEATURES  = agent_config['fixed_features']
TARGET_COL      = agent_config['target_col']
AGEGROUP_CONFIG = agent_config['agegroup_config']
NUM_FEATURES    = agent_config['num_features']

models = {}
for group_name in AGEGROUP_CONFIG:
    try:
        models[group_name] = joblib.load(
            f'../results/tables/model_{group_name}.pkl'
        )
    except FileNotFoundError:
        models[group_name] = None

with open('../results/tables/guardrail_ranges.json', 'r', encoding='utf-8') as f:
    all_guardrails = json.load(f)

guardrail_desc = pd.read_csv('../results/tables/guardrail_descriptive.csv')

print("Data loaded successfully.")

# %%
# ## 2. Kruskal-Wallis Test: Guardrail Range Heterogeneity

print("\n" + "="*60)
print("Section 2: Kruskal-Wallis Test — Guardrail Heterogeneity")
print("="*60)

KEY_VARS = ['BMI', 'Energy_kcal', 'Sodium_mg', 'Carb_g',
            'Sugar_g', 'WaistCirc', 'Weight']
GROUPS   = ['MiddleAged_Male', 'MiddleAged_Female',
            'Older_Male',      'Older_Female']

kw_records = []
for var in KEY_VARS:
    sub        = guardrail_desc[guardrail_desc['Variable'] == var]
    group_data = [sub[sub['Group'] == g]['Range_width'].values
                  for g in GROUPS]
    group_data = [g for g in group_data if len(g) > 0]
    if len(group_data) < 2:
        continue
    stat, p = kruskal(*group_data)
    kw_records.append({
        'Variable':    var,
        'H_statistic': round(stat, 4),
        'p_value':     round(p,    6),
        'Significant': 'Yes' if p < 0.05 else 'No',
    })
    print(f"  {var:20s}: H={stat:.4f}, p={p:.4f}  "
          f"{'*' if p < 0.05 else ''}")

kw_df = pd.DataFrame(kw_records)
kw_df.to_csv('../results/tables/kruskal_wallis_results.csv',
             index=False, encoding='utf-8-sig')
print(">>> Saved: ../results/tables/kruskal_wallis_results.csv")

# %%
# ## 3. Post-Hoc Pairwise Tests (Mann-Whitney U + Bonferroni)

print("\n" + "="*60)
print("Section 3: Post-Hoc Pairwise Tests")
print("="*60)

n_comp          = len(list(combinations(GROUPS, 2)))
alpha_corrected = 0.05 / n_comp
sig_vars        = kw_df[kw_df['Significant'] == 'Yes']['Variable'].tolist()

print(f"  Significant variables  : {sig_vars}")
print(f"  Bonferroni alpha       : {alpha_corrected:.4f} ({n_comp} pairs)")

dunn_records = []
for var in sig_vars:
    sub = guardrail_desc[guardrail_desc['Variable'] == var]
    for g1, g2 in combinations(GROUPS, 2):
        d1 = sub[sub['Group'] == g1]['Range_width'].values
        d2 = sub[sub['Group'] == g2]['Range_width'].values
        if len(d1) == 0 or len(d2) == 0:
            continue
        stat, p = mannwhitneyu(d1, d2, alternative='two-sided')
        p_corr  = min(p * n_comp, 1.0)
        dunn_records.append({
            'Variable':    var,
            'Group1':      g1,
            'Group2':      g2,
            'U_statistic': round(stat,   4),
            'p_raw':       round(p,      6),
            'p_corrected': round(p_corr, 6),
            'Significant': 'Yes' if p_corr < 0.05 else 'No',
        })

dunn_df = pd.DataFrame(dunn_records)
dunn_df.to_csv('../results/tables/dunn_posthoc.csv',
               index=False, encoding='utf-8-sig')

sig_pairs = dunn_df[dunn_df['Significant'] == 'Yes']
if len(sig_pairs) > 0:
    print(sig_pairs[['Variable', 'Group1', 'Group2',
                      'p_corrected']].to_string(index=False))
else:
    print("  No significant pairwise differences after Bonferroni correction.")
    print("  (Small n per group — 3 cases — limits pairwise power.)")
print(">>> Saved: ../results/tables/dunn_posthoc.csv")

# %%
# ## 4. Fallback Rate Analysis (Actual DiCE Search)

print("\n" + "="*60)
print("Section 4: Fallback Rate — Actual DiCE Search")
print("="*60)

def apply_hard_rules(query: pd.DataFrame, features: list) -> dict:
    cur = {f: float(query.iloc[0][f]) for f in features}
    g   = {}
    for f in features:
        g[f] = [round(max(cur[f] * 0.70, 0.0), 4),
                round(cur[f] * 1.30, 4)]
    for f in ['BMI', 'WaistCirc', 'Weight']:
        g[f] = [round(cur[f] * 0.85, 4), round(cur[f], 4)]
    wt_ratio = g['Weight'][0] / cur['Weight']
    g['WaistCirc'][0] = max(g['WaistCirc'][0],
                            round(cur['WaistCirc'] * wt_ratio, 4))
    g['BMI'][0]       = max(g['BMI'][0],
                            round(cur['BMI']       * wt_ratio, 4))
    g['Energy_kcal']  = [round(max(cur['Energy_kcal'] * 0.70, 500.0), 4),
                         round(cur['Energy_kcal'], 4)]
    g['Sodium_mg']    = [round(max(cur['Sodium_mg'] * 0.60, 800.0), 4),
                         round(cur['Sodium_mg'] * 0.90, 4)]
    g['Carb_g'][1]    = round(cur['Carb_g'],  4)
    g['Sugar_g'][1]   = round(cur['Sugar_g'], 4)
    g['Protein_g'][0]    = round(max(cur['Protein_g']    * 0.60, 30.0),  4)
    g['Potassium_mg'][0] = round(max(cur['Potassium_mg'] * 0.50, 500.0), 4)
    g['Carb_g'][0]       = round(max(cur['Carb_g']       * 0.20, 30.0),  4)
    g['Fiber_g'][0]      = round(max(cur['Fiber_g']       * 0.30, 5.0),  4)
    for f in features:
        if g[f][0] > g[f][1]:
            g[f] = [round(cur[f] * 0.85, 4), round(cur[f], 4)]
    return g

def build_search_range(hard_ranges, df_ref, features):
    sr = {}
    for feat in features:
        hard_lo, hard_hi = hard_ranges.get(feat, [0, float('inf')])
        data_lo = float(df_ref[feat].quantile(0.05))
        data_hi = float(df_ref[feat].quantile(0.95))
        lo = max(min(hard_lo, data_lo), 0.0)
        hi = max(hard_hi, data_hi)
        sr[feat] = [round(lo, 4), round(hi, 4)]
    return sr

fallback_records = []

for group_name, cfg in AGEGROUP_CONFIG.items():
    mdl = models.get(group_name)
    if mdl is None:
        continue

    age_grp = 0.0 if cfg['age_min'] < 60 else 1.0
    df_ref  = df_final[
        (df_final['AgeGroup'] == age_grp) &
        (df_final['Sex']      == cfg['sex_code'])
    ].copy()

    pool = df_final[
        (df_final['AgeGroup'] == age_grp) &
        (df_final['Sex']      == cfg['sex_code']) &
        (df_final[TARGET_COL] == 3.0) &
        (df_final['BMI']      >= 23.0)
    ].copy()

    preds     = mdl.predict(pool[X_FEATURES])
    confirmed = pool[preds == 3].sample(
        min(30, len(pool[preds == 3])), random_state=42
    )

    d   = dice_ml.Data(
            dataframe=df_final.copy().astype(float)[X_FEATURES + [TARGET_COL]],
            continuous_features=X_FEATURES,
            outcome_name=TARGET_COL,
          )
    m   = dice_ml.Model(model=mdl, backend="sklearn")
    exp = dice_ml.Dice(d, m, method="genetic")

    n_direct = n_fb1 = n_fb2 = n_fail = 0
    print(f"\n  [{group_name}] Testing {len(confirmed)} patients...")

    for _, row in confirmed.iterrows():
        query       = pd.DataFrame([row[X_FEATURES]])
        hard_ranges = apply_hard_rules(query, X_FEATURES)
        sr          = build_search_range(hard_ranges, df_ref, X_FEATURES)

        achieved = None
        for target_class in [0, 1, 2]:
            try:
                cf = exp.generate_counterfactuals(
                    query, total_CFs=1,
                    desired_class=target_class,
                    features_to_vary=VARY_FEATURES,
                    permitted_range=sr,
                    proximity_weight=0.2,
                    sparsity_weight=0.1,
                )
                if cf and len(cf.cf_examples_list[0].final_cfs_df) > 0:
                    achieved = target_class
                    break
            except Exception:
                continue

        if achieved == 0:   n_direct += 1
        elif achieved == 1: n_fb1    += 1
        elif achieved == 2: n_fb2    += 1
        else:               n_fail   += 1

    n_total = len(confirmed)
    fallback_records.append({
        'Group':        group_name,
        'N_tested':     n_total,
        'N_direct':     n_direct,
        'N_fallback1':  n_fb1,
        'N_fallback2':  n_fb2,
        'N_fail':       n_fail,
        'Direct_pct':   round(n_direct / n_total * 100, 2),
        'Fallback_pct': round((n_fb1 + n_fb2) / n_total * 100, 2),
        'Fail_pct':     round(n_fail  / n_total * 100, 2),
    })
    print(f"    Direct={n_direct}, FB1={n_fb1}, "
          f"FB2={n_fb2}, Fail={n_fail}")

fallback_df = pd.DataFrame(fallback_records)

# Chi-square: direct vs. non-direct
fallback_df['N_nondirect'] = (
    fallback_df['N_fallback1'] +
    fallback_df['N_fallback2'] +
    fallback_df['N_fail']
)
contingency = fallback_df[['N_direct', 'N_nondirect']].values
if contingency.min() == 0 or contingency.shape[0] < 2:
    print("\n  Chi-square skipped: insufficient variation in fallback counts.")
else:
    chi2, p_chi2, dof, _ = chi2_contingency(contingency)
    print(f"\n  Chi-square: χ²={chi2:.4f}, df={dof}, p={p_chi2:.4f}")
    fallback_df['Chi2_stat'] = round(chi2,   4)
    fallback_df['p_chi2']    = round(p_chi2, 6)

fallback_df.to_csv('../results/tables/fallback_analysis.csv',
                   index=False, encoding='utf-8-sig')
print(fallback_df[['Group', 'N_tested', 'Direct_pct',
                   'Fallback_pct', 'Fail_pct']].to_string(index=False))
print(">>> Saved: ../results/tables/fallback_analysis.csv")

# %%
# ## 5. Numerical Hallucination Comparison Table
# Generate Pure DiCE output for each representative case and compare
# against guardrail-constrained output to document hallucination patterns.

print("\n" + "="*60)
print("Section 5: Numerical Hallucination Comparison")
print("="*60)

DISPLAY_VARS = ['BMI', 'WaistCirc', 'Weight',
                'Energy_kcal', 'Carb_g', 'Sugar_g', 'Sodium_mg']

halluc_records = []

for case_key, gdata in all_guardrails.items():
    group_name  = gdata['group']
    mdl         = models.get(group_name)
    if mdl is None:
        continue

    cfg         = AGEGROUP_CONFIG[group_name]
    age_grp     = 0.0 if cfg['age_min'] < 60 else 1.0
    df_ref      = df_final[
        (df_final['AgeGroup'] == age_grp) &
        (df_final['Sex']      == cfg['sex_code'])
    ].copy()

    orig        = gdata['patient_profile']
    hard_ranges = gdata['final_ranges']
    query       = pd.DataFrame([orig])[X_FEATURES]

    df_stable = df_final.copy().astype(float)
    d   = dice_ml.Data(
            dataframe=df_stable[X_FEATURES + [TARGET_COL]],
            continuous_features=X_FEATURES,
            outcome_name=TARGET_COL,
          )
    m   = dice_ml.Model(model=mdl, backend="sklearn")
    exp = dice_ml.Dice(d, m, method="genetic")

    # Pure DiCE (no constraints)
    try:
        cf_pure = exp.generate_counterfactuals(
            query, total_CFs=4, desired_class=0,
            features_to_vary=VARY_FEATURES,
            proximity_weight=0.2, sparsity_weight=0.1,
        )
        pure_cfs = cf_pure.cf_examples_list[0].final_cfs_df.to_dict(
            orient='records'
        ) if cf_pure else []
    except Exception:
        pure_cfs = []

    # Guardrail DiCE (wide search + hard clip)
    sr = build_search_range(hard_ranges, df_ref, X_FEATURES)
    try:
        cf_grd = exp.generate_counterfactuals(
            query, total_CFs=4, desired_class=0,
            features_to_vary=VARY_FEATURES,
            permitted_range=sr,
            proximity_weight=0.2, sparsity_weight=0.1,
        )
        grd_cfs = cf_grd.cf_examples_list[0].final_cfs_df.to_dict(
            orient='records'
        ) if cf_grd else []
        # Apply hard clip
        for cf_row in grd_cfs:
            for feat, (lo, hi) in hard_ranges.items():
                if feat in cf_row:
                    cf_row[feat] = float(np.clip(float(cf_row[feat]), lo, hi))
    except Exception:
        grd_cfs = []

    # Identify worst hallucination in pure DiCE (most extreme BMI)
    if pure_cfs:
        worst_pure = min(pure_cfs, key=lambda r: float(r.get('BMI', 99)))
    else:
        worst_pure = {}

    # Pick most typical guardrail CF (closest to median)
    if grd_cfs:
        best_grd = grd_cfs[0]
    else:
        best_grd = {}

    # Check hallucination flags
    def bmi_weight_inconsistent(cf_row, orig_d):
        bmi_ratio = float(cf_row.get('BMI', orig_d['BMI'])) / float(orig_d['BMI'])
        wt_ratio  = float(cf_row.get('Weight', orig_d['Weight'])) / float(orig_d['Weight'])
        return abs(bmi_ratio - wt_ratio) > 0.05

    def energy_below_floor(cf_row, orig_d):
        floor = max(float(orig_d['Energy_kcal']) * 0.70, 500.0)
        return float(cf_row.get('Energy_kcal', orig_d['Energy_kcal'])) < floor

    def sodium_carb_conflict(cf_row, orig_d):
        na_cut   = float(cf_row.get('Sodium_mg', orig_d['Sodium_mg'])) < \
                   float(orig_d['Sodium_mg']) * 0.99
        carb_up  = float(cf_row.get('Carb_g',   orig_d['Carb_g']))   > \
                   float(orig_d['Carb_g'])   * 1.01
        return na_cut and carb_up

    rec = {
        'CaseKey': case_key,
        'Group':   group_name,
    }
    for var in DISPLAY_VARS:
        rec[f'Orig_{var}']  = round(float(orig.get(var, np.nan)), 2)
        rec[f'Pure_{var}']  = round(float(worst_pure.get(var, np.nan)), 2) \
                               if worst_pure else np.nan
        rec[f'Grd_{var}']   = round(float(best_grd.get(var, np.nan)), 2) \
                               if best_grd else np.nan

    rec['Pure_BMI_Weight_inconsistent'] = int(
        bmi_weight_inconsistent(worst_pure, orig)) if worst_pure else np.nan
    rec['Pure_Energy_below_floor']      = int(
        energy_below_floor(worst_pure, orig))      if worst_pure else np.nan
    rec['Pure_Conflict']                = int(
        sodium_carb_conflict(worst_pure, orig))    if worst_pure else np.nan
    rec['Grd_BMI_Weight_inconsistent']  = int(
        bmi_weight_inconsistent(best_grd, orig))   if best_grd  else np.nan
    rec['Grd_Energy_below_floor']       = int(
        energy_below_floor(best_grd, orig))        if best_grd  else np.nan
    rec['Grd_Conflict']                 = int(
        sodium_carb_conflict(best_grd, orig))      if best_grd  else np.nan

    halluc_records.append(rec)
    print(f"  [{case_key}]  "
          f"Pure BMI={rec.get('Pure_BMI', '?'):.2f}  "
          f"Pure Energy={rec.get('Pure_Energy_kcal', 0):.1f}  "
          f"Grd BMI={rec.get('Grd_BMI', '?'):.2f}")

halluc_df = pd.DataFrame(halluc_records)
halluc_df.to_csv('../results/tables/hallucination_comparison.csv',
                 index=False, encoding='utf-8-sig')

# Summary: violation counts
viol_cols_pure = ['Pure_BMI_Weight_inconsistent',
                  'Pure_Energy_below_floor', 'Pure_Conflict']
viol_cols_grd  = ['Grd_BMI_Weight_inconsistent',
                  'Grd_Energy_below_floor',  'Grd_Conflict']

print("\n[Hallucination Summary]")
print(f"  Pure DiCE violations (worst CF per case, n={len(halluc_df)}):")
for col in viol_cols_pure:
    cnt = halluc_df[col].sum()
    print(f"    {col:40s}: {int(cnt)}/{len(halluc_df)}")
print(f"  Guardrail violations (best CF per case, n={len(halluc_df)}):")
for col in viol_cols_grd:
    cnt = halluc_df[col].sum()
    print(f"    {col:40s}: {int(cnt)}/{len(halluc_df)}")
print(">>> Saved: ../results/tables/hallucination_comparison.csv")

# %%
# ## 6. Hallucination Comparison Figure

COMPARE_VARS = ['BMI', 'Energy_kcal', 'Sodium_mg']
var_labels   = ['BMI', 'Energy (kcal)', 'Sodium (mg)']
groups_plot  = list(AGEGROUP_CONFIG.keys())
colors       = {'Orig': '#7F8C8D', 'Pure': '#E74C3C', 'Grd': '#2980B9'}

fig, axes = plt.subplots(1, 3, figsize=(13, 5))

for ax, var, label in zip(axes, COMPARE_VARS, var_labels):
    x      = np.arange(len(groups_plot))
    w      = 0.25
    orig_means = []
    pure_means = []
    grd_means  = []

    for g in groups_plot:
        sub = halluc_df[halluc_df['Group'] == g]
        orig_means.append(sub[f'Orig_{var}'].mean())
        pure_means.append(sub[f'Pure_{var}'].mean())
        grd_means.append(sub[f'Grd_{var}'].mean())

    ax.bar(x - w, orig_means, w, label='Original',
           color=colors['Orig'], edgecolor='white', zorder=3)
    ax.bar(x,     pure_means, w, label='Pure DiCE (worst)',
           color=colors['Pure'], edgecolor='white', zorder=3)
    ax.bar(x + w, grd_means,  w, label='Guardrail DiCE',
           color=colors['Grd'],  edgecolor='white', zorder=3)

    ax.set_title(label, fontsize=11, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(
        ['MA\nMale', 'MA\nFemale', 'Old\nMale', 'Old\nFemale'],
        fontsize=9
    )
    ax.yaxis.grid(True, linestyle='--', alpha=0.4, zorder=0)
    ax.set_axisbelow(True)
    ax.spines[['top', 'right']].set_visible(False)
    if ax == axes[0]:
        ax.legend(fontsize=8)

fig.tight_layout()
save_fig(fig, 'figure_hallucination_comparison')
plt.show()
```
